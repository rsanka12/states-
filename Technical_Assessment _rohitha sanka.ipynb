```python
#import all the packages you think are needed
import pandas as pd 
import numpy as np 
import matplotlib as plt 
import seaborn as sns
```

```python
#Read in the Dataset into a pandas dataframe
df = pd.read_csv("Technical_Assessment_Data.csv")
```

```python
#Look at general statistics of the dataset and variables
print(df.describe())
```

```python
#Look at the first 8 rows of the dataset
print(df.head(8))
```

```python
#Look at the last 8 rows of the dataset
print(df.tail(8))
```

```python
#Drop the variables 'Urban', 'Unnamed: 0' and 'School Closure Date'
df = df[['State', 'Tested', 'Infected', 'Deaths', 'Population', 'Pop Density', 'Gini', 'ICU Beds', 'Income', 'GDP', 'Unemployment', 'Sex Ratio', 'Smoking Rate', 'Flu Deaths', 'Respiratory Deaths', 'Physicians', 'Hospitals', 'Health Spending', 'Pollution', 'Med-Large Airports', 'Temperature', 'Age 0-25', 'Age 26-54', 'Age 55+', 'Regions']]

# new df without urban, unamed, and school closure date
print(df.head())

```

```python
#Sort all states by Population in order of greatest to least
df_sorted = df.sort_values(by='Population', ascending=False)
print(df_sorted)
```

```python
#Group states by region and calculate mean for the variables
#'Population', 'Tested', and 'Infected'
region_means = df.groupby('Regions')[['Population', 'Tested', 'Infected']].mean()
print(region_means)
```

```python
#Create a new column that represents the Infected to Tested ratio that's called InfToTest
df['InfToTest'] = df['Infected'] / df['Tested']
print(df.head())
```

```python
#Make a scatter plot with the variables 'Tested' and 'Infected'
#Draw a conclusion from the graph you made
sns.scatterplot(data=df, x="Tested", y="Infected")
plt.pyplot.title("Scatter Plot of Tested vs Infected")
plt.pyplot.xlabel("Tested")
plt.pyplot.ylabel("Infected")
plt.pyplot.show()
#Conclusion: The scatterplot I designed represents the relationship between the number of tests conducted by the X axis and the number
# of confirmed infections in each state. The relationship between the two variables 
#is not completely linear, the scatter plot identifies outliers and also potentional C19 testing gaps.
```

```python
#Output all of the states with a population of over 10 million (10000000) and less than a GDP of 65 thousand (65000)
#How many states fit this criteria?
filtered_states = df[(df['Population'] > 10000000) & (df['GDP'] < 65000)]
print(filtered_states)
num_states = filtered_states.shape[0]
print("Number of states that fit the criteria:", num_states)
#number of states that fit this criteria is 7
```

```python
#Output all the states that are below the average population and are above average in infected OR have a pollution index of
#less than 5
avg_population = df['Population'].mean()
avg_infected = df['Infected'].mean()
filtered_states = df[(df['Population'] < avg_population) & ((df['Infected'] > avg_infected) | (df['Pollution'] < 5))]
print(filtered_states)
```

```python
#Create a linear regression model with two variables of your choice
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data = pd.read_csv('Technical_Assessment_Data.csv')
# Select features (Smoking Rate) and target variable (Respiratory Deaths)
X = data[['Smoking Rate']]
y = data['Respiratory Deaths']
model = LinearRegression()
model.fit(X, y)
plt.scatter(X, y, color='blue')
plt.xlabel('Smoking Rate')
plt.ylabel('Respiratory Deaths')
x_values = np.linspace(X.min(), X.max(), 100)
y_values = model.predict(x_values.reshape(-1, 1))
plt.plot(x_values, y_values, color='red')
plt.title('Smoking Rate vs. Respiratory Deaths')
plt.show()
#I built a regression model comparing smoking rate to respiratory deaths. We see an incline in both as 
#people who smoke a lot also are victims of respiratory deaths
```
